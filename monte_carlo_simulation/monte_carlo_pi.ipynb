```python
# Monte Carlo Simulation to Estimate π

import numpy as np
import matplotlib.pyplot as plt
```

```python

# 1) Generate random points
N = 100_000
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)
```

```python
# 2) Check which points fall inside the circle
inside = x**2 + y**2 <= 1
pi_estimate = 4 * np.sum(inside) / N
print(f"Monte Carlo π estimate with {N:,} samples: {pi_estimate:.6f}")
```

```python
# Visualize the first 2000 points
plt.figure(figsize=(6,6))
plt.scatter(x[:2000], y[:2000], c=inside[:2000], cmap="bwr", s=5)
circle = plt.Circle((0,0), 1, color='black', fill=False, linewidth=2)
plt.gca().add_patch(circle)
plt.gca().set_aspect('equal')
plt.title("Monte Carlo Simulation of π (First 2000 Points)")
plt.show()
```

```python
# Convergence plot: how estimate approaches π
estimates = []
for i in range(1, N+1):
    if inside[i-1]:
        count_inside = np.sum(inside[:i])
    else:
        count_inside = np.sum(inside[:i])
    estimates.append(4 * count_inside / i)

plt.figure(figsize=(8,5))
plt.plot(estimates, linewidth=1)
plt.axhline(np.pi, color='red', linestyle='--', label="True π")
plt.xlabel("Number of samples")
plt.ylabel("π estimate")
plt.title("Convergence of Monte Carlo π Estimate")
plt.legend()
plt.show()
```

📌 Conclusion

The estimate of π improves as N increases.

Error decreases at a rate of 1/√N (law of large numbers).

This demonstrates how Monte Carlo turns randomness into accuracy.
